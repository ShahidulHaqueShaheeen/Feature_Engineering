```python
# Finding the IQR
percentile25 = df.placement_exam_marks.quantile(0.25)
percentile75 = df.placement_exam_marks.quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
upper_b = percentile75 + (1.5 * iqr)
lower_b = percentile25 - (1.5 * iqr)
```

```python
upper_b
```

```python
lower_b
```

# Finding Outliers

```python
df[(df.placement_exam_marks < lower_b) | (df.placement_exam_marks > upper_b)].shape
```

# Trimming

```python
new_df = df[(df.placement_exam_marks > lower_b) & (df.placement_exam_marks < upper_b)]
new_df
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

# Capping the outliers

```python
df.placement_exam_marks = np.where(df.placement_exam_marks > upper_b,upper_b,np.where(df.placement_exam_marks < lower_b,lower_b,df.placement_exam_marks))
```

```python
sns.distplot(df.placement_exam_marks)
```

```python
sns.boxplot(data = df,x = df.placement_exam_marks)
```

**So, we can see that, outliers are capped**
